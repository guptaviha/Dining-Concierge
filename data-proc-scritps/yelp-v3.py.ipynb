```python
import boto3
from decimal import Decimal
import json
import os
import requests
import time
```

```python
dynamodb = boto3.resource('dynamodb')
table = dynamodb.Table('yelp-restaurants-v2')

def save_in_dynamo(id_, name, address, cuisine, rating):
    table.put_item(
       Item={
            'id': id_,
            'name': name,
            'address': address,
            'cuisine': cuisine,
            'rating': Decimal(rating),
            'insertedAtTimestamp': int(time.time())
        }
    )
    return

curl = 'curl -XPUT -u yelp:Yelp@123 \'https://search-yelp-restaurant-data-idw5yplh7blonn57r3fi3prbv4.us-east-1.es.amazonaws.com/restaurants/_doc/'

def save_in_es(id_, cuisine):
    s = curl + id_ + '\' -d \'' + json.dumps({'id':id_,'cuisine':cuisine}) + '\' -H \'Content-Type: application/json\''
    os.popen(s).read()
    return
```

```python
base_path = '/Users/aatman/nyu/Sem2/CC/assignment1/data/manhattan/'
cuisines = ['chinese', 'indian', 'italian', 'mexican', 'thai']

cuisine = 'thai'
o = open("/Users/aatman/nyu/Sem2/CC/assignment1/" + cuisine + ".txt", "w")
count = 0
file_path = base_path + cuisine + "/"
for filename in os.listdir(file_path):
    f = open(file_path + filename)
    data = json.load(f)
    f.close()
    id_ = data['id']
    name = data['name']
    address = ''
    for add in data['location']['display_address']:
        address += add + ', '
    address = address[:len(address)-2]
    rating = data['rating']
    save_in_dynamo(id_, name, address, cuisine, rating)
    save_in_es(id_, cuisine)
    o.write(id_ + '\t' + name + '\t' + address + '\t' + cuisine)
    count += 1
    if count % 100 == 0:
        print(count)
        time.sleep(5)
print(cuisine, "--", count)

o.close()
```
